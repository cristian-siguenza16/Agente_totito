```python
import random
import math
import os
```

```python
class Jugador:
    def __init__(self,simbolo):
        self.simbolo = simbolo  #Letra del jugador (X o O)
    
    #posiciona el simbolo en lugar dado 
    def posicionar(self,tablero):
        while True:
            posicion =  int(input("Ingrese numero de posicion para el simbolo de (1-9): "))  # del 1 al 9 horizontalmente
            print()
            if tablero[posicion-1] == "-":
                break
        return posicion-1
```

```python
class totito:
    def __init__(self):
      #crea los 3 sombolos principales del juego
      #vacio jugador x y jugador o, para el humano y la ia de manera aletaoria
        self.tablero = ['-' for _ in range(9)]
        if random.randint(0, 1) == 1:
            self.jugador = 'X'
            self.ia = "O"
        else:
            self.jugador = "O"
            self.ia = "X"

    def mostrarTablero(self):
        print("")
        for i in range(3):
            print("  ",self.tablero[0+(i*3)]," | ",self.tablero[1+(i*3)]," | ",self.tablero[2+(i*3)])
            print("")
            
    def tableroLleno(self,espacio):
        return not "-" in espacio

    def condicionVictoria(self,tablero,jugador):
        if tablero[0]==tablero[1]==tablero[2] == jugador: return True
        if tablero[3]==tablero[4]==tablero[5] == jugador: return True
        if tablero[6]==tablero[7]==tablero[8] == jugador: return True
        if tablero[0]==tablero[3]==tablero[6] == jugador: return True
        if tablero[1]==tablero[4]==tablero[7] == jugador: return True
        if tablero[2]==tablero[5]==tablero[8] == jugador: return True
        if tablero[0]==tablero[4]==tablero[8] == jugador: return True
        if tablero[2]==tablero[4]==tablero[6] == jugador: return True
        #todas las rayas posibles del juego
        return False

    def comprobarGanador(self):
        if self.condicionVictoria(self.tablero,self.jugador):
            os.system("cls")
            print(f"   Gana {self.jugador}")
            return True
            
        if self.condicionVictoria(self.tablero,self.ia):
            os.system("cls")
            print(f"   Gana {self.ia}")
            return True

     
        if self.tableroLleno(self.tablero):
            os.system("cls")
            print("   Empate!")
            return True
        return False

    def inicio(self):
        ia = inteligenciaA(self.ia)
        jugador = Jugador(self.jugador)
        while True:
            os.system("cls")
            print(f"{self.jugador} turno jugador...")
            self.mostrarTablero()
            
            #Jugador
            posicion = jugador.posicionar(self.tablero)
            self.tablero[posicion] = self.jugador
            if self.comprobarGanador():
                break
            
            #Bot
            posicion = ia.movimientoIA(self.tablero)
            self.tablero[posicion] = self.ia
            if self.comprobarGanador():
                break

        #El tablero que se ve al final
        print()
        self.mostrarTablero()
```

```python
class inteligenciaA(totito):
    def __init__(self,simbolo):
        self.ia = simbolo
        self.jugador = "X" if simbolo == "O" else "O"

    def asignarTurno(self,tablero):
      #Comprobamos como va el juego 
      #A su vez comom van los turnos del mismo
        x = 0
        o = 0
        for i in range(9):
            if(tablero[i] == "X"):
                x = x+1
            if(tablero[i] == "O"):
                o = o+1
        
        if(self.jugador == "X"):
            return "X" if x==o else "O"
        if(self.jugador == "O"):
            return "O" if x==o else "X"
    
    def posicionesVacias(self,tablero):
        return [i for i, x in enumerate(tablero) if x == '-']
    
    def cambioTablero(self,tablero,jugada):
        #crea el tablero con una jugada nueva
        nuevoTablero = tablero.copy()
        jugador = self.asignarTurno(tablero)
        nuevoTablero[jugada] = jugador
        return nuevoTablero
    
    def predecir(self,tablero):
       #Intenta predecir un posible 3 en raya
        if(self.condicionVictoria(tablero,"X")):
            return True
        if(self.condicionVictoria(tablero,"O")):
            return True
        return False

    def minimax(self, tablero, jugando):
        jugadorH = self.jugador #Jugador humano "Falso"
        jugadorIA = 'O' if jugando == 'X' else 'X' #Jugador IA

        
        if self.predecir(tablero):#if que predice si hay una victoria segura 
            return {'Posicion': None, 'Puntos': 1 * (len(self.posicionesVacias(tablero)) + 1) if jugadorIA == jugadorH else 
                    -1 * (len(self.posicionesVacias(tablero)) + 1)}
        elif self.tableroLleno(tablero):#La partida sigue en curso y en empate
            return {'Posicion': None, 'Puntos': 0}

        if jugando == jugadorH:
            mejor = {'Posicion': None, 'Puntos': -math.inf}  # Minima puntuacion(Dejar al humano con la menor puntuacionn)
        else:
            mejor = {'Posicion': None, 'Puntos': math.inf}  # Maxima puntuacion(PAra que l maquina pueda ganar)
            #En este se ven todas las jugadas posibles crando tableros
        for movimientosPosibles in self.posicionesVacias(tablero):
            nuevoTablero = self.cambioTablero(tablero,movimientosPosibles)
            puntuacion = self.minimax(nuevoTablero, jugadorIA)  # Simla el juego despues del movimiento

            puntuacion['Posicion'] = movimientosPosibles  # Es el movimiento mas optimo
            if jugando == jugadorH:  
                if puntuacion['Puntos'] > mejor['Puntos']:
                    mejor = puntuacion
            else:
                if puntuacion['Puntos'] < mejor['Puntos']:
                    mejor = puntuacion
        return mejor # Se regresa la mejor opcion para el 
    def movimientoIA(self,tablero):
        posicion = self.minimax(tablero,self.ia)['Posicion']
        return posicion
```
